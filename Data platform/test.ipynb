```python
import html
```

```python
data['TEN_THU_TUC'] = data['TEN_THU_TUC'].replace(';', '.',regex=True)
data['TEN_THU_TUC'] = data['TEN_THU_TUC'].apply(html.unescape)

data['TEN_LINH_VUC_THU_TUC'] = data['TEN_LINH_VUC_THU_TUC'].replace(';', '.',regex=True)
data['TEN_LINH_VUC_THU_TUC'] = data['TEN_LINH_VUC_THU_TUC'].apply(html.unescape)

data['TEN_LOAI_CONG_VIEC'] = data['TEN_LOAI_CONG_VIEC'].replace(';', '.',regex=True)
data['TEN_LOAI_CONG_VIEC'] = data['TEN_LOAI_CONG_VIEC'].apply(html.unescape)

data['TEN_DON_VI'] = data['TEN_DON_VI'].replace(';', '.',regex=True)
data['TEN_DON_VI'] = data['TEN_DON_VI'].apply(html.unescape)

data['TEN_TO_CHUYEN_MON'] = data['TEN_TO_CHUYEN_MON'].replace(';', '.',regex=True)
data['TEN_TO_CHUYEN_MON'] = data['TEN_TO_CHUYEN_MON'].apply(html.unescape)

data['TEN_CAN_BO'] = data['TEN_CAN_BO'].replace(';', '.',regex=True)
data['TEN_CAN_BO'] = data['TEN_CAN_BO'].apply(html.unescape)
```

```python
data[:1000].to_csv("C:/Users/Admin/Postman/files/ds_danhsachcongviechoso_clean_test.csv", sep="\t", index=False)
```
